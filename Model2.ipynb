```python
import numpy as np
import tellurium as te
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
model = te.loada('''
    compartment C1;
    C1 = 100;
    species TCE, Epoxide, Dichloroactate, LossOfOneCL, Glyoxylate, Output;

    TCE = 1.0;
    Epoxide = 0;
    Dichloroactate = 0;
    LossOfOneCL = 0;
    Glyoxylate = 0;
    Output = 0;
    
    TCE in C1; Epoxide in C1; Dichloroactate in C1; LossOfOneCL in C1; Glyoxylate in C1; Output in C1;
    
    J1: TCE -> Epoxide; k2*TCE*PopulationVariable1; 
    J2: Epoxide -> Dichloroactate; k3*Epoxide/4;
    J3: Dichloroactate -> LossOfOneCL; k4*Dichloroactate*PopulationVariable2;
    J4: LossOfOneCL -> Glyoxylate; PopulationVariable3*k4*LossOfOneCL;
    J5: Glyoxylate -> Output; k5*Glyoxylate*PopulationVariable3*k1

    k1 = .00045;
    k2 = .4;
    k3 = .49;
    k4 = 545;
    k5 = 500;
    
    PopulationVariable1=50;
    PopulationVariable2=50;
    PopulationVariable3=50;

''')
```

```python
model.integrator.setValue('maximum_num_steps', 2000000)
```

```python
def Population1(x):
    model.reset()
    model.PopulationVariable1 = x
    model.simulate(0, 45, 1000)
    z = model.plot()
    model.show(z)
    
def Population2(x):
    model.reset()
    model.PopulationVariable2 = x
    model.simulate(0, 45, 1000)
    z = model.plot()
    model.show(z)

def Population3(x):
    model.reset()
    model.PopulationVariable3 = x
    model.simulate(0, 45, 1000)
    z = model.plot()
    model.show(z)
    

interact(Population1, x=(0, 100));
interact(Population2, x=(0, 100));
interact(Population3, x=(0, 100));
```

```python
model.reset()
result = model.gillespie (0, 45, 1000, ['time', 'TCE', 'Epoxide', 'Dichloroactate', 'LossOfOneCL', 'Glyoxylate', 'Output' ])
model.plot(result)
```

```python
model.draw(width=300)
model.conservedMoietyAnalysis = True
model.plot(result);
```
